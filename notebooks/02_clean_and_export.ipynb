```python
import pandas as pd

df = pd.read_csv('../data/raw/WA_Fn-UseC_-HR-Employee-Attrition.xls', encoding='utf-8-sig')
```

```python
constant_cols = [col for col in df.columns if df[col].nunique() == 1]
print("상수 컬럼:", constant_cols)
```

```python
print(df[constant_cols].head(1))
```

```python
df_cleaned = df.drop(columns=constant_cols)

print("원본:", df.shape, 
      "정제 후:", df_cleaned.shape)
print(df_cleaned.columns.tolist())
```

```python
df_cleaned.to_csv('../data/processed/hr_cleaned.csv', index=False, encoding='utf-8-sig')
print("저장 완료:", df_cleaned.shape)
```
