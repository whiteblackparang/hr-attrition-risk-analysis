```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('../sql/hr_analysis.db')

def run_sql_file(filepath, conn):
    with open(filepath, 'r', encoding='utf-8') as f:
        sql_script = f.read()
    
    queries = [q.strip() for q in sql_script.split(';') if q.strip()]
    
    results = []
    for i, q in enumerate(queries, 1):
        clean_q = '\n'.join(line for line in q.split('\n') if not line.strip().startswith('--'))
        if not clean_q.strip():
            continue
        print(f"--- Query {i} ---")
        result = pd.read_sql(clean_q, conn)
        print(result)
        print()
        results.append(result)
    return results
```
